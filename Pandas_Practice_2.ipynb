```python
ipl = pd.read_csv('/content/ipl-matches.csv')
```

```python
ipl.head(2)
```

```python
#which player has won the most potm --> in finals and qauliffiers
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
#Toss decision plot
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
#how many matches teams have played

Most_Playing_Teams = (ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False)
```

```python
Most_Playing_Teams.plot(kind='bar')
```
